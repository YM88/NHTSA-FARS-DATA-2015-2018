```python
import numpy as np
import pandas as pd
```

```python
from data_processing import process_data

df = process_data(2018)
df
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100, max_depth=10)

x = df[["lon", "lat", "vehicles", "persons", "fatals", "drunk_dr", "reckless", "impaired"]]
y = df["speeding"]

clf.fit(x, y)
```

```python
print(clf.feature_importances_)
```

```python
clf.score(x, y)
```
